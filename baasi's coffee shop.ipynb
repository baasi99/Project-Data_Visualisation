```python
empty_values_count = data.replace('', pd.NA).isna().sum()
empty_values_count
```

```python
data_by_month = data.groupby(['Month Name']).agg({'Total_Bill': ['sum', 'mean', 'max', 'min'], 'transaction_id':['count']}).dropna()
data_by_month
```

```python
months_ordered = ['January', 'February', 'March', 'April', 'May', 'June',
                  'July', 'August', 'September', 'October', 'November', 'December']
data['Month Name'] = pd.Categorical(data['Month Name'], categories=months_ordered, ordered=True)
monthly_sales = data.groupby('Month Name', observed=True).sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Month Name', y='Total_Bill', data=monthly_sales, order=monthly_sales['Month Name'], color='skyblue')
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## -IF we are taking last six month sales , June is the peak

```python
daily_data = data.groupby('Day Name', observed=True)['Total_Bill'].sum().reset_index()
plt.figure(figsize=(9, 5))
sns.lineplot(x='Day Name', y='Total_Bill', data=daily_data, color='red', marker='o')
plt.title('Daily Sales Data')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## -If we are taking daily basis , Tuesday is getting more sale

```python
store_wise_sales = data.groupby('store_location', observed=True)['Total_Bill'].sum().reset_index()
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='store_location', y='Total_Bill', data=store_wise_sales, order=store_wise_sales['store_location'], color='skyblue')
plt.title('Sales Data By Store')
plt.xlabel('Store')
plt.ylabel('Total Sales')

```

## Currently Astoria is the store which is on High sale

```python
sales_summary = data.groupby('product_detail')['transaction_qty'].sum().reset_index()
sales_summary = sales_summary.sort_values(by='transaction_qty', ascending=False)
plt.figure(figsize=(10, 6))
plt.pie(sales_summary['transaction_qty'], labels=sales_summary['product_detail'], autopct='%1.1f%%', startangle=140,shadow=True,explode = (0.1, 0, 0, 0,0,0,0,0))
plt.title('Best-Selling Products')
plt.axis('equal') 
plt.show()
```

## Best selling product - English breakfast

```python
heatmap_data = data.pivot_table(values='transaction_qty', 
                                  index='Day of Week', 
                                  columns='Hour', 
                                  aggfunc='sum', 
                                  fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Transaction Quantities Heat Map')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
```
